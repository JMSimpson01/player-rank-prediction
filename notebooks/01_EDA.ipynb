```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(
    r"C:\Users\jmsim\OneDrive\Desktop\player-rank-prediction\data\raw\chess_games.csv"
)

print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
def categorize_elo(elo):
    if elo < 1000:
        return "Beginner"
    elif elo < 1400:
        return "Intermediate"
    elif elo < 1800:
        return "Advanced"
    else:
        return "Expert"

df['skill_level'] = df['white_rating'].apply(categorize_elo)

print(df['skill_level'].value_counts())
```

```python
df['skill_level'].value_counts().plot(kind='bar')

plt.title("Chess Player Skill Distribution")
plt.xlabel("Skill Level")
plt.ylabel("Number of Players")

plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True
)

plt.title("Feature Correlation Heatmap")

plt.show()
```
